```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dtypes #datatype of data
```

```python
#clean up data
```

```python
#Only want USA RACES 50K or 50 miles
```

```python
#step 1, show 50m or 50k
#50k and 50mi
```

```python
df[df['Event distance/length'] == '50mi']
```

```python
#combine 50k and 50mi with isin
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020)]
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine all filters together
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')
]
```

```python
df2.head(10)
```

```python
#remove (USA) From event name
```

```python
df2['Event name']= df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
df2.head()
```

```python
#remove h from athlete performance 
df2['Athlete performance']= df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns Athlete club, Athlete coountry, Athlete dob,Athlete age category,  
```

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete club','Athlete age category','Athlete year of birth'], axis = 1)
```

```python
df2.head()
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
df2.reset_index(drop = True)
```

```python
df2['athlete_age'] = df2['athlete_age'].astype('int')
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete club                  object
#Athlete country               object
#Athlete year of birth        float64
#Athlete gender                object
#Athlete age category          object
#Athlete average speed         object
#Athlete ID                     int64
#dtype: object
```

```python
df2 = df2.rename(columns = {
    'Year of event':'Year',
    'Event dates': 'race_day',
    'Event name':'race_name',
    'Event distance/length':'race_distance',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed' : 'athlete_average_speed',
    'Athlete ID' : 'athlete_id'   
})
```

```python
df2.head()
```

```python
#reorder columns
```

```python
df3 = df2[['race_day','race_name','race_distance','race_number_of_finishers', 'athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
df3.head()   
          
```

```python
df3[df3['race_name'] == 'West Seattle Beach Run - Winter Edition (USA)']
```

```python
sns.histplot(df3['race_distance'])
```

```python
sns.histplot(df3, x = 'race_distance', hue = 'athlete_gender')
```

```python
sns.distplot(df3[df3['race_distance']== '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(df3 , x = 'race_distance', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart',linewidth=1)
```

```python
sns.lmplot(data=df3, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
#quesetions from the data that can be answered
```

```python
#difference in speed for 50km, 50mi races. 


df3.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()


```

```python
#what age groups are the worst in 50mi race?
df3.groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>20')
```

```python
#seasons for data
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head()
```

```python
#all races by seasons
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
#race in mi by season
df3.query('race_distance == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
#race in km by seasons
df3.query('race_distance == "50km"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
